## Hyperparameter: Number of Recurrent Layers

### 1. Number of Recurrent Layers = 1

```python
layer1Suite = RNNSuite("layer1.keras", "layer1.json")
print("[INFO] Loading datasets...")
df_train, df_val, df_test = layer1Suite.load_dataset()

print("[INFO] Preprocessing text...")
vectorizer, x_train, x_val, x_test = layer1Suite.preprocess_text(df_train, df_val, df_test)

print("[INFO] Encoding labels...")
label_encoder, y_train, y_val, y_test = layer1Suite.encode_labels(df_train, df_val, df_test)

vocab_size = len(vectorizer.get_vocabulary())
print(f"[INFO] Vocab size: {vocab_size}")

print("[INFO] Training Keras RNN model...")
layer1Suite.train_rnn(x_train, y_train, x_val, y_val, x_test, y_test, vocab_size, rnn_layers=1)

print("[INFO] Running ScratchRNN forward pass...")
layer1Suite.evaluate_scratch_model(x_test, y_test)

print("[START] Plotting training history...")
layer1Suite.plot_history()
```
