## Hyperparameter: Number of Convolutional Layers

### 1. Number of Conv Layers = 1

```python
cnn_convLayer_1 = CNNSuite(save_dir="saved_models", model_name="cnn_convLayers_1.keras")
cnn_convLayer_1.history_path = cnn_convLayer_1.save_dir / "cnn_convLayers_1_history.json"
cnn_convLayer_1.load_data()

if cnn_convLayer_1.model_path.exists() and cnn_convLayer_1.history_path.exists():
    print("[INFO] Model and history already exist. Skipping training and loading history...")
    cnn_convLayer_1.load_history()
else:
    cnn_convLayer_1.train(epochs=10, batch_size=64, conv_layers=1)

keras_cnn_convLayer_1 = cnn_convLayer_1.evaluate_keras()
scratch_cnn_convLayer_1 = cnn_convLayer_1.evaluate_scratch()
cnn_convLayer_1.plot_history()
print(f"[RESULT] Conv Layers = 1 | Keras F1 = {keras_cnn_convLayer_1:.6f} | Scratch F1 = {scratch_cnn_convLayer_1:.6f}\n")
```

## Hyperparameter: Pooling Type (Max / Avg)

### 1. Max Pooling

```python
cnn_max = CNNSuite(save_dir="saved_models", model_name="cnn_pooling_max.keras")
cnn_max.history_path = cnn_max.save_dir / "cnn_pooling_max_history.json"
cnn_max.load_data()

if cnn_max.model_path.exists() and cnn_max.history_path.exists():
    print("[INFO] Pooling = max: Model and history already exist. Skipping training...")
    cnn_max.load_history()
else:
    cnn_max.train(epochs=10, batch_size=64, pooling="max")

keras_cnn_max = cnn_max.evaluate_keras()
scratch_cnn_max = cnn_max.evaluate_scratch()
cnn_max.plot_history()
print(f"[RESULT] Pooling = Max | Keras F1 = {keras_cnn_max:.6f} | Scratch F1 = {scratch_cnn_max:.6f}\n")
```

### 2. Average Pooling

```python
cnn_avg = CNNSuite(save_dir="saved_models", model_name="cnn_pooling_avg.keras")
cnn_avg.history_path = cnn_avg.save_dir / "cnn_pooling_avg_history.json"
cnn_avg.load_data()

if cnn_avg.model_path.exists() and cnn_avg.history_path.exists():
    print("[INFO] Pooling = avg: Model and history already exist. Skipping training...")
    cnn_avg.load_history()
else:
    cnn_avg.train(epochs=10, batch_size=64, pooling="avg", pool_size=2)

keras_cnn_avg = cnn_avg.evaluate_keras()
scratch_cnn_avg = cnn_avg.evaluate_scratch()
cnn_avg.plot_history()
print(f"[RESULT] Pooling = Average | Keras F1 = {keras_cnn_avg:.6f} | Scratch F1 = {scratch_cnn_avg:.6f}\n")
```

```python
histories_pool = {
    "max pooling": cnn_max.load_history(),
    "avg pooling": cnn_avg.load_history(),
}

plot_multi_loss(histories_pool)
```
