## Hyperparameter: Number of Convolutional Layers

### 1. Number of Conv Layers = 1

```python
suite1 = CNNSuite(save_dir="saved_models", model_name="cnn_convLayers_1.keras")
suite1.history_path = suite1.save_dir / "cnn_convLayers_1_history.json"
suite1.load_data()
suite1.train(epochs=10, batch_size=64)
keras_f1_1 = suite1.evaluate_keras()
scratch_f1_1 = suite1.evaluate_scratch()
suite1.plot_history()
print(f"[RESULT] Conv Layers = 1 | Keras F1 = {keras_f1_1:.5f} | Scratch F1 = {scratch_f1_1:.5f}\n")
```

### 2. Number of Conv Layers = 2

```python
suite2 = CNNSuite(save_dir="saved_models", model_name="cnn_convLayers_2.keras")
suite2.history_path = suite2.save_dir / "cnn_convLayers_2_history.json"
suite2.load_data()
suite2.train(epochs=10, batch_size=64)
keras_f1_2 = suite2.evaluate_keras()
scratch_f1_2 = suite2.evaluate_scratch()
suite2.plot_history()
print(f"[RESULT] Conv Layers = 2 | Keras F1 = {keras_f1_2:.4f} | Scratch F1 = {scratch_f1_2:.4f}\n")
```
